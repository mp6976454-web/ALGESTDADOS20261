```python
#Maior entre Dois
#Leia dois números inteiros. Exiba qual deles é o maior. Se forem iguais, exiba "Os números são iguais."
num1 = int(input("digite numero"))
num2 = int(input("digite outro numero"))
if num1 > num2:
    print(f"Maior do que, {num2} ")
elif num1 < num2:
    print(f"Maior do que, {num1}")
else:
    print(f"os numeros são iguais")
```
