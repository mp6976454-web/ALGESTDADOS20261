```
#09 PESQUISAR
temperatura = float(input("digite temperatura"))
unidade_origem = input('kelvin: K - Celsius: C - Fahrenheit: F')

temp_celsius = ''
temp_kelvin = ''
temp_fahrenheit = ''

if temperatura < 0 and unidade_origem == 'K':
    print("valor inválido - valores para kelvin não pode ser negativo")

else:
    match unidade_origem:
        case 'K':
            temp_celsius = temperatura - 273
            temp_fahrenheit = (temperatura-273)*1.8 + 32
            print(f'A temperatura {temperatura} em Kelvin quando covertida para demais escalas (celsius/fahreinheit) fica \n{temp_celsius} em celsius \n{temp_fahrenheit:.2f} em fahrenheit')
        case 'C':
            temp_fahrenheit = (temperatura*1.8) + 32
            temp_kelvin = temperatura - 273
            print(f'A temperatura {temperatura} em celsius quando covertida para demais escalas (kelvin/fahreinheit) fica \n{temp_kelvin} em kelvin \n{temp_fahrenheit} em fahrenheit')

        case 'F':
            temp_celsius = (temperatura - 32)/1.8
            temp_kelvin = (temperatura - 32) * (5/9) + 273
            print(f'A temperatura {temperatura} em fahrenheit quando covertida para demais escalas (kelvin/celsius) fica \n{temp_celsius} em celsius \n{temp_kelvin:.2f} em kelvin')

        case _:
          print('unidade inválida')
```

```
#13
preco_original = float(input("Digite o preço original do produto: R$"))
codigo_cupom_input = input("Digite o código do cupom (PROMO10, PROMO20, PROMO30, SEMCUPOM): ").upper()

desconto_base_percentual = 0.0
cupom_usado_msg = ""

match codigo_cupom_input:
    case "PROMO10":
        desconto_base_percentual = 0.10
        cupom_usado_msg = "PROMO10"
    case "PROMO20":
        desconto_base_percentual = 0.20
        cupom_usado_msg = "PROMO20"
    case "PROMO30":
        desconto_base_percentual = 0.30
        cupom_usado_msg = "PROMO30"
    case "SEMCUPOM":
        desconto_base_percentual = 0.0
        cupom_usado_msg = "SEMCUPOM"
    case _:
        print("Código de cupom inválido. Nenhum desconto será aplicado.")
        cupom_usado_msg = "Nenhum válido"

desconto_aplicado_percentual = desconto_base_percentual
desconto_cap_mensagem = ""

if preco_original < 50:
    if desconto_base_percentual > 0.10:
        desconto_aplicado_percentual = 0.10
        desconto_cap_mensagem = "(limitado a 10% devido ao preço abaixo de R$50,00)"
    else:
        desconto_aplicado_percentual = desconto_base_percentual
elif preco_original >= 50:
    desconto_aplicado_percentual = desconto_base_percentual

desconto_em_reais = preco_original * desconto_aplicado_percentual
preco_final = preco_original - desconto_em_reais

print(f"\n--- Detalhes da Compra ---")
print(f"Preço Original: R${preco_original:.2f}")
print(f"Cupom Usado: {cupom_usado_msg} {desconto_cap_mensagem}")
print(f"Desconto em Reais: R${desconto_em_reais:.2f}")
print(f"Preço Final: R${preco_final:.2f}")
```
