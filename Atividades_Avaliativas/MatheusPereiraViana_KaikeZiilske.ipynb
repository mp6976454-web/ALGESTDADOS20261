```
# EXERCÍCIO 16 - Simulador de Bomba de Combustível

print("\nEXERCÍCIO 16 - Simulador de Bomba de Combustível")
tipo_combustivel = input('Digite o combustível ("gasolina", "etanol", "diesel"): ').lower()
# Input validation for litros
while True:
    try:
        litros = float(input("Digite a quantidade de litros: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

preco_gasolina = 5.79
preco_etanol = 3.89
preco_diesel = 6.19

if litros <= 0:
    print("Quantidade inválida.")
else:
    match tipo_combustivel:
        case "gasolina":
            preco_litro = preco_gasolina
        case "etanol":
            preco_litro = preco_etanol
        case "diesel":
            preco_litro = preco_diesel
        case _:
            preco_litro = None

    if preco_litro is None:
        print("Tipo de combustível inválido.")
    else:
        subtotal = litros * preco_litro

        if litros > 40:
            desconto = subtotal * 0.03
        else:
            desconto = 0

        total = subtotal - desconto

        print(f"Subtotal: R$ {subtotal:.2f}")
        print(f"Desconto: R$ {desconto:.2f}")
        print(f"Total a pagar: R$ {total:.2f}")
```

```
# EXERCÍCIO 17 - Classificador de Velocidade

print("\nEXERCÍCIO 17 - Classificador de Velocidade")
# Input validation for velocidade
while True:
    try:
        velocidade = float(input("Digite a velocidade em km/h: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")
tipo_via = input('Digite o tipo da via ("urbana", "rodovia", "autopista"): ').lower()

if velocidade < 0:
    print("Velocidade inválida.")
else:
    match tipo_via:
        case "urbana":
            limite = 60
        case "rodovia":
            limite = 100
        case "autopista":
            limite = 120
        case _:
            limite = None

    if limite is None:
        print("Tipo de via inválido.")
    else:
        excesso = velocidade - limite

        if velocidade <= limite:
            situacao = "Dentro do limite"
        elif velocidade <= limite * 1.20:
            situacao = "Excesso leve"
        elif velocidade <= limite * 1.50:
            situacao = "Excesso moderado"
        else:
            situacao = "Excesso grave"

        print(f"Limite da via: {limite} km/h")
        print(f"Velocidade informada: {velocidade:.2f} km/h")
        print(f"Situação: {situacao}")
```

```
# EXERCÍCIO 18 - Calculadora de Parcelas com Juros

print("\nEXERCÍCIO 18 - Calculadora de Parcelas com Juros")
# Input validation for valor_compra
while True:
    try:
        valor_compra = float(input("Digite o valor da compra: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for parcelas
while True:
    try:
        parcelas = int(input("Digite o número de parcelas (1 a 12): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if valor_compra <= 0 or parcelas < 1 or parcelas > 12:
    print("Entrada inválida.")
else:
    if parcelas == 1:
        juros = 0.00
    elif parcelas <= 3:
        juros = 0.05
    elif parcelas <= 6:
        juros = 0.10
    else:
        juros = 0.15

    valor_total = valor_compra * (1 + juros)
    valor_parcela = valor_total / parcelas

    print(f"Valor total com juros: R$ {valor_total:.2f}")
    print(f"Valor de cada parcela: R$ {valor_parcela:.2f}")

    if valor_parcela < 20:
        print("Aviso: parcela mínima de R$20,00 não atingida.")
        parcelas_minimas = int(valor_total // 20)
        if valor_total % 20 != 0:
            parcelas_minimas += 1
        print(f"Número mínimo de parcelas para respeitar R$20,00: {parcelas_minimas}")
```

```
# EXERCÍCIO 19 - Detector de Intervalo com Tolerância

print("\nEXERCÍCIO 19 - Detector de Intervalo com Tolerância")
# Input validation for alvo
while True:
    try:
        alvo = float(input("Digite o valor alvo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for tolerancia
while True:
    try:
        tolerancia = float(input("Digite a tolerância: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for valor_medido
while True:
    try:
        valor_medido = float(input("Digite o valor medido: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if alvo <= 0 or tolerancia < 0 or valor_medido <= 0:
    print("Entrada inválida.")
else:
    limite_inferior = alvo - tolerancia
    limite_superior = alvo + tolerancia
    desvio = valor_medido - alvo

    if limite_inferior <= valor_medido <= limite_superior:
        classificacao = "Dentro da faixa"
    elif valor_medido < limite_inferior:
        classificacao = "Abaixo do limite"
    else:
        classificacao = "Acima do limite"

    print(f"Limite inferior: {limite_inferior:.2f}")
    print(f"Limite superior: {limite_superior:.2f}")
    print(f"Desvio: {desvio:.2f}")
    print(f"Classificação: {classificacao}")
```

```
# EXERCÍCIO 21 - Calculadora de Energia Elétrica

print("\nEXERCÍCIO 21 - Calculadora de Energia Elétrica")
# Input validation for consumo
while True:
    try:
        consumo = float(input("Digite o consumo do mês em kWh: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if consumo < 0:
    print("Consumo inválido.")
else:
    if consumo <= 100:
        preco_kwh = 0.50
    elif consumo <= 200:
        preco_kwh = 0.75
    elif consumo <= 300:
        preco_kwh = 1.00
    else:
        preco_kwh = 1.50

    valor_sem_desconto = consumo * preco_kwh
    valor_com_desconto = valor_sem_desconto

    if consumo > 300:
        modo_economico = input("Optou pelo modo econômico? (s/n): ").lower()
        if modo_economico == "s":
            valor_com_desconto = valor_sem_desconto * 0.90

    print(f"Valor sem desconto: R$ {valor_sem_desconto:.2f}")
    print(f"Valor com desconto: R$ {valor_com_desconto:.2f}")
```

```
# EXERCÍCIO 22 - Verificador de Intervalo Múltiplo

print("\nEXERCÍCIO 22 - Verificador de Intervalo Múltiplo")
# Input validation for A
while True:
    try:
        A = float(input("Digite o intervalo mínimo A: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for B
while True:
    try:
        B = float(input("Digite o intervalo máximo B: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for N
while True:
    try:
        N = int(input("Digite o número N: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for D
while True:
    try:
        D = int(input("Digite o divisor D: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if A >= B or D == 0 or D < 0:
    print("Entrada inválida.")
else:
    dentro_intervalo = A <= N <= B
    divisivel = N % D == 0

    if dentro_intervalo and divisivel:
        print("Dentro do intervalo e divisível.")
    elif dentro_intervalo and not divisivel:
        print("Dentro do intervalo e não divisível.")
    elif not dentro_intervalo and divisivel:
        print("Fora do intervalo e divisível.")
    else:
        print("Fora do intervalo e não divisível.")
```

```
# EXERCÍCIO 23 - Sistema de Pontuação de Jogo

print("\nEXERCÍCIO 23 - Sistema de Pontuação de Jogo")
# Input validation for p1
while True:
    try:
        p1 = int(input("Pontuação do jogador 1: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for p2
while True:
    try:
        p2 = int(input("Pontuação do jogador 2: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

# Input validation for p3
while True:
    try:
        p3 = int(input("Pontuação do jogador 3: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")

if (p1 < 0 or p1 > 1000 or p2 < 0 or p2 > 1000 or p3 < 0 or p3 > 1000):
    print("Pontuação inválida.")
else:
    if p1 >= p2 and p1 >= p3:
        primeiro = p1
        nome1 = "Jogador 1"
        if p2 >= p3:
            segundo = p2
            nome2 = "Jogador 2"
            terceiro = p3
            nome3 = "Jogador 3"
        else:
            segundo = p3
            nome2 = "Jogador 3"
            terceiro = p2
            nome3 = "Jogador 2"
    elif p2 >= p1 and p2 >= p3:
        primeiro = p2
        nome1 = "Jogador 2"
        if p1 >= p3:
            segundo = p1
            nome2 = "Jogador 1"
            terceiro = p3
            nome3 = "Jogador 3"
        else:
            segundo = p3
            nome2 = "Jogador 3"
            terceiro = p1
            nome3 = "Jogador 1"
    else:
        primeiro = p3
        nome1 = "Jogador 3"
        if p1 >= p2:
            segundo = p1
            nome2 = "Jogador 1"
            terceiro = p2
            nome3 = "Jogador 2"
        else:
            segundo = p2
            nome2 = "Jogador 2"
            terceiro = p1
            nome3 = "Jogador 1"

    diferenca = primeiro - terceiro

    print(f"1º lugar: {nome1} com {primeiro} pontos")
    print(f"2º lugar: {nome2} com {segundo} pontos")
    print(f"3º lugar: {nome3} com {terceiro} pontos")
    print(f"Diferença entre 1º e 3º: {diferenca} pontos")
```

```
# EXERCÍCIO 24 - Calculadora de Reajuste Salarial

print("\nEXERCÍCIO 24 - Calculadora de Reajuste Salarial")
# Input validation for salario_atual
while True:
    try:
        salario_atual = float(input("Digite o salário atual: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for avaliacao
while True:
    try:
        avaliacao = float(input("Digite o percentual de avaliação de desempenho (0 a 100): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if salario_atual <= 0 or avaliacao < 0 or avaliacao > 100:
    print("Entrada inválida.")
else:
    if avaliacao >= 90:
        percentual = 0.15
    elif avaliacao >= 75:
        percentual = 0.10
    elif avaliacao >= 60:
        percentual = 0.05
    else:
        percentual = 0.00

    reajuste = salario_atual * percentual
    novo_salario = salario_atual + reajuste

    if novo_salario > 15000:
        novo_salario = 15000
        reajuste = novo_salario - salario_atual
        print("Aviso: reajuste limitado ao teto de R$15.000,00.")

    print(f"Salário anterior: R$ {salario_atual:.2f}")
    print(f"Percentual aplicado: {percentual * 100:.1f}%")
    print(f"Reajuste em reais: R$ {reajuste:.2f}")
    print(f"Novo salário: R$ {novo_salario:.2f}")
```

```
# EXERCÍCIO 25 - Classificador de Triângulo com Ângulos

print("\nEXERCÍCIO 25 - Classificador de Triângulo com Ângulos")
# Input validation for angulo1
while True:
    try:
        angulo1 = float(input("Digite o primeiro ângulo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for angulo2
while True:
    try:
        angulo2 = float(input("Digite o segundo ângulo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for angulo3
while True:
    try:
        angulo3 = float(input("Digite o terceiro ângulo: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if angulo1 <= 0 or angulo2 <= 0 or angulo3 <= 0:
    print("Ângulos inválidos.")
else:
    soma = angulo1 + angulo2 + angulo3

    if soma != 180:
        print(f"Triângulo inválido. Soma encontrada: {soma}")
    else:
        if angulo1 < 90 and angulo2 < 90 and angulo3 < 90:
            tipo_angulo = "acutângulo"
        elif angulo1 == 90 or angulo2 == 90 or angulo3 == 90:
            tipo_angulo = "retângulo"
        else:
            tipo_angulo = "obtusângulo"

        if angulo1 == angulo2 == angulo3:
            tipo_lado = "equilátero"
        elif angulo1 == angulo2 or angulo1 == angulo3 or angulo2 == angulo3:
            tipo_lado = "isósceles"
        else:
            tipo_lado = "escaleno"

        print(f"Triângulo {tipo_angulo} e {tipo_lado}.")
```

```
# EXERCÍCIO 26 - Calculadora de Combustível de Viagem

print("\nEXERCÍCIO 26 - Calculadora de Combustível de Viagem")
# Input validation for distancia
while True:
    try:
        distancia = float(input("Digite a distância da viagem (km): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for consumo
while True:
    try:
        consumo = float(input("Digite o consumo do veículo (km/l): "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

# Input validation for preco_litro
while True:
    try:
        preco_litro = float(input("Digite o preço do combustível por litro: R$ "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

if distancia <= 0 or consumo <= 0 or preco_litro <= 0:
    print("Entrada inválida.")
else:
    litros_necessarios = distancia / consumo
    custo_total = litros_necessarios * preco_litro

    print(f"Litros necessários: {litros_necessarios:.2f} L")
    print(f"Custo total: R$ {custo_total:.2f}")

    if custo_total > 500:
        parada = distancia / 2
        print(f"Custo ultrapassa R$500,00. Sugestão de parada para abastecimento no km {parada:.2f}.")
```

```
# EXERCÍCIO 27 - Verificador de Senha com Nível de Segurança

print("\nEXERCÍCIO 27 - Verificador de Senha com Nível de Segurança")
senha = input("Digite a senha: ")
confirmacao = input("Confirme a senha: ")

if senha != confirmacao:
    print("Senhas não conferem.")
else:
    tamanho = len(senha)

    if tamanho < 6:
        nivel = "Fraca"
        nivel_num = 1
    elif tamanho <= 9:
        nivel = "Média"
        nivel_num = 2
    else:
        nivel = "Forte"
        nivel_num = 3

    if senha.isdigit():
        nivel_num -= 1
        if nivel_num < 1:
            nivel_num = 1

    if nivel_num == 1:
        nivel_final = "Fraca"
    elif nivel_num == 2:
        nivel_final = "Média"
    else:
        nivel_final = "Forte"

    print(f"Nível de segurança: {nivel_final}")
```

```
# EXERCÍCIO 29 - Simulador de Batalha por Tipos

print("\nEXERCÍCIO 29 - Simulador de Batalha por Tipos")
tipo_atacante = input('Digite o tipo do atacante ("fogo", "agua", "planta", "eletrico", "terra"): ').lower()
tipo_defensor = input('Digite o tipo do defensor ("fogo", "agua", "planta", "eletrico", "terra"): ').lower()
# Input validation for poder_base
while True:
    try:
        poder_base = float(input("Digite o poder base do ataque: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um valor numérico válido.")

tipos_validos = ["fogo", "agua", "planta", "eletrico", "terra"]

if tipo_atacante not in tipos_validos or tipo_defensor not in tipos_validos or poder_base <= 0:
    print("Entrada inválida.")
else:
    vantagem = (
        (tipo_atacante == "fogo" and tipo_defensor == "planta") or
        (tipo_atacante == "agua" and tipo_defensor == "fogo") or
        (tipo_atacante == "planta" and tipo_defensor == "agua") or
        (tipo_atacante == "eletrico" and tipo_defensor == "agua") or
        (tipo_atacante == "terra" and tipo_defensor == "eletrico")
    )

    desvantagem = (
        (tipo_defensor == "fogo" and tipo_atacante == "planta") or
        (tipo_defensor == "agua" and tipo_atacante == "fogo") or
        (tipo_defensor == "planta" and tipo_atacante == "agua") or
        (tipo_defensor == "eletrico" and tipo_atacante == "agua") or
        (tipo_defensor == "terra" and tipo_atacante == "eletrico")
    )

    if vantagem:
        multiplicador = 2
        resultado = "vantagem"
    elif desvantagem:
        multiplicador = 0.5
        resultado = "desvantagem"
    else:
        multiplicador = 1
        resultado = "neutro"

    dano_final = poder_base * multiplicador

    print(f"Situação: {resultado}")
    print(f"Dano final: {dano_final:.2f}")
```

```
# EXERCÍCIO 30 - Sistema Completo de Reserva de Hotel

print("\nEXERCÍCIO 30 - Sistema Completo de Reserva de Hotel")
acomodacao = input('Digite a acomodação ("standard", "luxo", "suite"): ').lower()
# Input validation for diarias
while True:
    try:
        diarias = int(input("Digite o número de diárias: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro válido.")
fidelidade = input("Possui cartão fidelidade? (s/n): ").lower()

diaria_standard = 180.00
diaria_luxo = 320.00
diaria_suite = 550.00

if diarias <= 0 or fidelidade not in ["s", "n"]:
    print("Entrada inválida.")
else:
    match acomodacao:
        case "standard":
            diaria_base = diaria_standard
        case "luxo":
            diaria_base = diaria_luxo
        case "suite":
            diaria_base = diaria_suite
        case _:
            diaria_base = None

    if diaria_base is None:
        print("Tipo de acomodação inválido.")
    else:
        subtotal = diaria_base * diarias

        if 1 <= diarias <= 3:
            desconto_qtd = 0.00
        elif 4 <= diarias <= 7:
            desconto_qtd = 0.10
        else:
            desconto_qtd = 0.20

        valor_com_desconto_qtd = subtotal * (1 - desconto_qtd)

        if fidelidade == "s":
            desconto_fidelidade = 0.05
        else:
            desconto_fidelidade = 0.00

        total_final = valor_com_desconto_qtd * (1 - desconto_fidelidade)

        if total_final < 100:
            total_final = 100

        print(f"Resumo da reserva:")
        print(f"Acomodação: {acomodacao}")
        print(f"Diária base: R$ {diaria_base:.2f}")
        print(f"Número de diárias: {diarias}")
        print(f"Subtotal: R$ {subtotal:.2f}")
        print(f"Desconto por quantidade: {desconto_qtd * 100:.0f}%")
        print(f"Desconto fidelidade: {desconto_fidelidade * 100:.0f}%")
        print(f"Total final: R$ {total_final:.2f}")
```
