```python
#Crie um programa que peça uma temperatura em graus Celsius e converta para Fahrenheit.

temperatura = float(input("Digite_a_Temperatura_em_graus"))
conversão = ({temperatura}*1,8) +32
print(f"a temperatura em fahrenheit é {conversão}")
```
